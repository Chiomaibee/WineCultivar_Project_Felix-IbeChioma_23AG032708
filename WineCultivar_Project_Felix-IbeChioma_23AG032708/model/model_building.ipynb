```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import joblib
import os

# 1. Load Dataset
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['cultivar'] = wine.target

# 2. Feature Selection (Selecting exactly 6 features)
features = ['alcohol', 'malic_acid', 'ash', 'magnesium', 'flavanoids', 'color_intensity']
X = df[features]
y = df['cultivar']

# 3. Feature Scaling (Mandatory)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Train Model (Random Forest)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 5. Evaluate
print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
print(classification_report(y_test, model.predict(X_test)))

# 6. Save Model and Scaler
if not os.path.exists('model'): os.makedirs('model')
joblib.dump(model, 'model/wine_cultivar_model.pkl')
joblib.dump(scaler, 'model/wine_scaler.pkl')
print("Model saved successfully as wine_cultivar_model.pkl")
```
